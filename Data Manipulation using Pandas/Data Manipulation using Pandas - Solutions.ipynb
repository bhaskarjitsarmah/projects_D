#### Q2. Create a separate dataframe with the columns `winner`, `total`, `voters` and save them to variable results

```python
results = election[['winner', 'total', 'voters']]
results.head()
```

#### Q3. Slice the dataframe using index row labels from `Perry` to `Potter` and then save them to variable `p_counties`

```python
p_counties = election['Perry':'Potter']
p_counties.head()
```

#### Q4. Slice the dataframe using index row labels from `Perry` to `Potter` in reverse order and then save them to variable `p_counties_rev`

```python
p_counties_rev = election['Potter':'Perry':-1]
p_counties_rev.head()
```

#### Q5. Slice the dataframe columns from the starting column to `Obama` and save them to variable `left_columns`

```python
left_columns = election.loc[:, 'state':'Obama']
left_columns.head()
```

#### Q6. Slice the dataframe columns from `Romney` to the end and save them to variable `right_columns`

```python
right_columns = election.loc[:, 'Romney':]
right_columns.head()
```

#### Q7. Prepare a boolean array to select all of the rows and columns where voter `turnout` exceeded 70% and save it variable `high_turnout_df`

```python
high_turnout_df = election[election['turnout'] > 70]
high_turnout_df.head()
```

#### Q8.  Convert the rows of the `winner` column to have missing values where `margin` column is less than 1%
